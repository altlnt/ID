```python
# Plot gyro, mag, baro, gps alt
```

```python
figure(figsize=(12,12))

gcf().add_subplot(411)

plot(data.t,data['gyro_x'],label="x",color="blue")
plot(data.t,data['gyro_y'],label="y",color="orange")
plot(data.t,data['gyro_z'],label="z",color="green")
gca().set_ylabel('rad/sec')

grid()
legend()
title('Gyro')

gcf().add_subplot(412)
plot(data.t,data['gyro_x'],label="x",color="blue")
grid()
legend()
gca().set_ylabel('rad/sec')

gcf().add_subplot(413)
plot(data.t,data['gyro_y'],label="y",color="orange")
grid()
legend()
gca().set_ylabel('rad/sec')
gcf().add_subplot(414)
plot(data.t,data['gyro_z'],label="z",color="green")
grid()
legend()
gca().set_xlabel('t (s)')
```

```python
figure(figsize=(12,12))
plot(data.t,data['mag_x'],label="x")
plot(data.t,data['mag_y'],label="y")
plot(data.t,data['mag_z'],label="z")
grid()
legend()
title('Mag')
gca().set_xlabel('t (s)')
gca().set_ylabel('G')
```

```python
figure(figsize=(11,11))
N=sqrt(sum([data[i]**2 for i in data.keys() if 'mag' in i],axis=0))
print([i for i in data.keys() if 'mag' in i])
plot(data.t,N,label="norm")
grid()
legend()
title('Mag norm')
gca().set_xlabel('t (s)')
gca().set_ylabel('G')
```

```python
figure(figsize=(7,7))
plot(data.t,data['accelero_x'],label="x")
plot(data.t,data['accelero_y'],label="y")
plot(data.t,data['accelero_z'],label="z")
grid()
legend()
title('Acc')
gca().set_xlabel('t (s)')
gca().set_ylabel('m/$s^2$')
```

```python
figure(figsize=(7,7))
N=sqrt(sum([data[i]**2 for i in data.keys() if 'accelero' in i],axis=0))
print([i for i in data.keys() if 'accelero' in i])
plot(data.t,N,label="norm")
grid()
legend()
title('Acc norm')
gca().set_xlabel('t (s)')
gca().set_ylabel('m/$s^2$')
```

```python
figure(figsize=(12,12))
suptitle('Vel')

gcf().add_subplot(311)

plot(data.t,data['vx_iner'],label="vx_ekf")
plot(data.t,data['v_gps_north'],label="vnorth_gps")

grid()
legend()

gcf().add_subplot(312)

plot(data.t,data['vy_iner'],label="vy_ekf")
plot(data.t,data['v_gps_east'],label="veast_gps")

grid()
legend()

gcf().add_subplot(313)

plot(data.t,data['vz_iner'],label="vz_ekf")
plot(data.t,data['v_gps_down'],label="vdown_gps")


grid()
legend()
gca().set_xlabel('t (s)')
gca().set_ylabel('m/s')
```

```python
figure(figsize=(12,12))
suptitle('Alt')

gcf().add_subplot(111)

plot(data.t,data['baro']-data['baro'].values[0],label="baro")
plot(data.t,0.001*(data['alt']-data['alt'].values[0]),label="gps")
plot(data.t,-cumsum(data.dt*data['v_gps_down']),label="integrated gps vdown")
grid()
legend()
```

```python
figure(figsize=(12,12))
suptitle('Airspeed')

gcf().add_subplot(111)

plot(data.t,data['airspeed_mes'],label="measured airspeed")

grid()
legend()
```
